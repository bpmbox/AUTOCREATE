## ğŸš€ Step 3: OCRé€£æºRPAã‚·ã‚¹ãƒ†ãƒ ã®åˆæœŸåŒ–

```python
# Colabç’°å¢ƒã§ã¯å®Ÿéš›ã®VNCæ“ä½œã¯ã§ããªã„ãŸã‚ã€ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³ç‰ˆã‚’ä½¿ç”¨
if IN_COLAB:
    print("ğŸŒŸ Colabç’°å¢ƒï¼šã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³ãƒ¢ãƒ¼ãƒ‰ã§å®Ÿè¡Œ")
    
    class ColabOCRRPASimulator:
        """Google Colabç”¨ã®OCR RPA ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚¿ãƒ¼"""
        
        def __init__(self, config):
            self.config = config
            self.simulation_data = {
                "screenshots": [
                    {"filename": "homepage.png", "page_type": "homepage", "content": "é‡‘ãƒ»è²´é‡‘å±ãƒ»ãƒ–ãƒ©ãƒ³ãƒ‰å“è²·å– kinkaimasu.jp"},
                    {"filename": "assessment.png", "page_type": "assessment_page", "content": "ç„¡æ–™æŸ»å®šãƒ•ã‚©ãƒ¼ãƒ  ãŠåå‰ é›»è©±ç•ªå· ãƒ¡ãƒ¼ãƒ«ã‚¢ãƒ‰ãƒ¬ã‚¹"},
                    {"filename": "prices.png", "page_type": "category_page", "content": "é‡‘ 1g 8,500å†† ãƒ—ãƒ©ãƒãƒŠ 1g 4,200å††"}
                ],
                "current_index": 0
            }
        
        def simulate_intelligent_analysis(self, iteration):
            """çŸ¥çš„è§£æã®ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³"""
            data_index = iteration % len(self.simulation_data["screenshots"])
            current_data = self.simulation_data["screenshots"][data_index]
            
            return {
                "screenshot_path": current_data["filename"],
                "detected_page_type": current_data["page_type"],
                "found_prices": ["8,500å††", "4,200å††"] if "å††" in current_data["content"] else [],
                "found_buttons": ["ç”³è¾¼", "é€ä¿¡"] if "ãƒ•ã‚©ãƒ¼ãƒ " in current_data["content"] else ["ã‚¯ãƒªãƒƒã‚¯"],
                "ocr_text": current_data["content"],
                "confidence": 0.9,
                "recommended_actions": [
                    {"type": "click_text", "target_text": "ç”³è¾¼", "description": "ç”³è¾¼ãƒœã‚¿ãƒ³ã‚¯ãƒªãƒƒã‚¯", "priority": 9},
                    {"type": "scroll", "direction": "down", "description": "ãƒšãƒ¼ã‚¸ã‚¹ã‚¯ãƒ­ãƒ¼ãƒ«", "priority": 7}
                ]
            }
        
        def run_simulation_loop(self, max_iterations=5):
            """ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³ãƒ«ãƒ¼ãƒ—å®Ÿè¡Œ"""
            results = {
                "iterations": [],
                "total_screenshots": 0,
                "successful_actions": 0,
                "collected_data": [],
                "goal_achieved": False
            }
            
            for i in range(max_iterations):
                print(f"ğŸ”„ ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³åå¾© {i+1}/{max_iterations}")
                
                analysis = self.simulate_intelligent_analysis(i)
                
                print(f"ğŸ“¸ ã‚¹ã‚¯ãƒªãƒ¼ãƒ³ã‚·ãƒ§ãƒƒãƒˆ: {analysis['screenshot_path']}")
                print(f"ğŸ§  ãƒšãƒ¼ã‚¸ã‚¿ã‚¤ãƒ—: {analysis['detected_page_type']}")
                print(f"ğŸ’° ç™ºè¦‹ä¾¡æ ¼: {analysis['found_prices']}")
                
                # ãƒ‡ãƒ¼ã‚¿åé›†
                if analysis['found_prices']:
                    results['collected_data'].extend([
                        {"type": "price", "value": price, "iteration": i+1}
                        for price in analysis['found_prices']
                    ])
                
                results['iterations'].append({
                    "iteration": i+1,
                    "analysis": analysis,
                    "success": True
                })
                
                results['total_screenshots'] += 1
                results['successful_actions'] += len(analysis['recommended_actions'])
                
                if len(results['collected_data']) >= 4:
                    results['goal_achieved'] = True
                    print("ğŸ‰ ç›®æ¨™é”æˆï¼")
                    break
            
            return results
    
    # ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚¿ãƒ¼åˆæœŸåŒ–
    automation_system = ColabOCRRPASimulator(config)
    print("âœ… Colabã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚¿ãƒ¼åˆæœŸåŒ–å®Œäº†")
    
else:
    print("ğŸ–¥ï¸ ãƒ­ãƒ¼ã‚«ãƒ«ç’°å¢ƒï¼šå®Ÿéš›ã®VNCæ“ä½œã‚·ã‚¹ãƒ†ãƒ ã‚’åˆæœŸåŒ–")
    # å®Ÿéš›ã®OCR RPA ã‚·ã‚¹ãƒ†ãƒ ã‚’ã‚¤ãƒ³ãƒãƒ¼ãƒˆãƒ»åˆæœŸåŒ–
    try:
        from scripts.ocr_rpa_automation import OCRRPAAutomation
        automation_system = OCRRPAAutomation(
            vnc_container=config["vnc_settings"]["container_name"],
            gas_ocr_url=config["gas_ocr_api"]["url"]
        )
        print("âœ… å®Ÿæ©Ÿã‚·ã‚¹ãƒ†ãƒ åˆæœŸåŒ–å®Œäº†")
    except ImportError as e:
        print(f"âš ï¸  ãƒ­ãƒ¼ã‚«ãƒ«ã‚·ã‚¹ãƒ†ãƒ åˆæœŸåŒ–ã‚¨ãƒ©ãƒ¼: {e}")
        print("ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³ãƒ¢ãƒ¼ãƒ‰ã§ç¶šè¡Œ...")
        automation_system = ColabOCRRPASimulator(config)

print("\nğŸ¯ ã‚·ã‚¹ãƒ†ãƒ æº–å‚™å®Œäº†ï¼")
```

## ğŸ§  Step 4: çŸ¥çš„è‡ªå‹•åŒ–ãƒ«ãƒ¼ãƒ—ã®å®Ÿè¡Œ

### OCRè§£æ â†’ ã‚¢ã‚¯ã‚·ãƒ§ãƒ³æ±ºå®š â†’ å®Ÿè¡Œã®ãƒ«ãƒ¼ãƒ—

```python
import time
from datetime import datetime

print("ğŸš€ AUTOCREATEçŸ¥çš„è‡ªå‹•åŒ–ãƒ«ãƒ¼ãƒ—é–‹å§‹")
print("="*50)

# è‡ªå‹•åŒ–ãƒ«ãƒ¼ãƒ—å®Ÿè¡Œ
start_time = time.time()

if hasattr(automation_system, 'run_simulation_loop'):
    # Colabã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚·ãƒ§ãƒ³
    results = automation_system.run_simulation_loop(max_iterations=5)
else:
    # å®Ÿæ©Ÿå®Ÿè¡Œ
    results = automation_system.run_intelligent_automation_loop(
        max_iterations=5,
        target_goal="data_collection"
    )

end_time = time.time()
execution_time = end_time - start_time

print("\n" + "="*50)
print("ğŸ“Š å®Ÿè¡Œçµæœã‚µãƒãƒªãƒ¼")
print("="*50)
print(f"â±ï¸  å®Ÿè¡Œæ™‚é–“: {execution_time:.2f}ç§’")
print(f"ğŸ”„ åå¾©å›æ•°: {len(results['iterations'])}å›")
print(f"ğŸ“¸ ã‚¹ã‚¯ãƒªãƒ¼ãƒ³ã‚·ãƒ§ãƒƒãƒˆ: {results['total_screenshots']}æš")
print(f"âœ… æˆåŠŸã‚¢ã‚¯ã‚·ãƒ§ãƒ³: {results['successful_actions']}ä»¶")
print(f"ğŸ’ åé›†ãƒ‡ãƒ¼ã‚¿: {len(results['collected_data'])}ä»¶")
print(f"ğŸ¯ ç›®æ¨™é”æˆ: {'âœ… ã¯ã„' if results['goal_achieved'] else 'âŒ ã„ã„ãˆ'}")

# åé›†ãƒ‡ãƒ¼ã‚¿ã®è©³ç´°è¡¨ç¤º
if results['collected_data']:
    print("\nğŸ’° åé›†ã—ãŸä¾¡æ ¼ãƒ‡ãƒ¼ã‚¿:")
    for data in results['collected_data']:
        print(f"  ğŸ”¹ åå¾©{data['iteration']}: {data['value']}")
```

## ğŸ“ˆ Step 5: çµæœã®å¯è¦–åŒ–

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# æ—¥æœ¬èªãƒ•ã‚©ãƒ³ãƒˆè¨­å®š
plt.rcParams['font.family'] = 'DejaVu Sans'
sns.set_style("whitegrid")

# çµæœãƒ‡ãƒ¼ã‚¿ã‚’DataFrameã«å¤‰æ›
iterations_data = []
for iteration in results['iterations']:
    iterations_data.append({
        'iteration': iteration['iteration'],
        'success': iteration.get('success', False),
        'page_type': iteration.get('analysis', {}).get('detected_page_type', 'unknown'),
        'found_prices_count': len(iteration.get('analysis', {}).get('found_prices', [])),
        'confidence': iteration.get('analysis', {}).get('confidence', 0)
    })

df_iterations = pd.DataFrame(iterations_data)

# å¯è¦–åŒ–
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('ğŸ›ï¸ AUTOCREATE OCR-RPA Automation Results', fontsize=16, fontweight='bold')

# 1. åå¾©ã”ã¨ã®æˆåŠŸç‡
axes[0, 0].bar(df_iterations['iteration'], df_iterations['success'].astype(int), color='skyblue')
axes[0, 0].set_title('ğŸ”„ Iteration Success Rate')
axes[0, 0].set_xlabel('Iteration')
axes[0, 0].set_ylabel('Success (1=True, 0=False)')

# 2. ãƒšãƒ¼ã‚¸ã‚¿ã‚¤ãƒ—åˆ†å¸ƒ
page_type_counts = df_iterations['page_type'].value_counts()
axes[0, 1].pie(page_type_counts.values, labels=page_type_counts.index, autopct='%1.1f%%', startangle=90)
axes[0, 1].set_title('ğŸŒ Detected Page Types')

# 3. ä¾¡æ ¼ãƒ‡ãƒ¼ã‚¿ç™ºè¦‹æ•°
axes[1, 0].plot(df_iterations['iteration'], df_iterations['found_prices_count'], marker='o', linewidth=2, color='green')
axes[1, 0].set_title('ğŸ’° Price Data Discovery')
axes[1, 0].set_xlabel('Iteration')
axes[1, 0].set_ylabel('Prices Found')
axes[1, 0].grid(True, alpha=0.3)

# 4. OCRä¿¡é ¼åº¦
axes[1, 1].bar(df_iterations['iteration'], df_iterations['confidence'], color='orange', alpha=0.7)
axes[1, 1].set_title('ğŸ¯ OCR Confidence Score')
axes[1, 1].set_xlabel('Iteration')
axes[1, 1].set_ylabel('Confidence')
axes[1, 1].set_ylim(0, 1)

plt.tight_layout()
plt.show()

# ãƒ‡ãƒ¼ã‚¿ãƒ†ãƒ¼ãƒ–ãƒ«è¡¨ç¤º
print("\nğŸ“‹ è©³ç´°ãƒ‡ãƒ¼ã‚¿ãƒ†ãƒ¼ãƒ–ãƒ«:")
print(df_iterations.to_string(index=False))
```

## ğŸ¯ Step 6: ãƒ“ã‚¸ãƒã‚¹ä¾¡å€¤ã®åˆ†æ

```python
# ãƒ“ã‚¸ãƒã‚¹ä¾¡å€¤è¨ˆç®—
manual_time_per_page = 30  # æ‰‹å‹•ã§1ãƒšãƒ¼ã‚¸è§£æã™ã‚‹æ™‚é–“ï¼ˆç§’ï¼‰
automation_time_per_page = execution_time / len(results['iterations'])  # è‡ªå‹•åŒ–ã§ã®1ãƒšãƒ¼ã‚¸è§£ææ™‚é–“

time_saving = manual_time_per_page - automation_time_per_page
efficiency_gain = (time_saving / manual_time_per_page) * 100

# æœˆé–“ä½œæ¥­é‡ã§ã®è¨ˆç®—
pages_per_month = 1000  # æœˆé–“è§£æãƒšãƒ¼ã‚¸æ•°ï¼ˆä»®å®šï¼‰
monthly_time_saving_hours = (time_saving * pages_per_month) / 3600
hourly_wage = 2000  # æ™‚çµ¦ï¼ˆå††ï¼‰
monthly_cost_saving = monthly_time_saving_hours * hourly_wage

print("ğŸ’¼ AUTOCREATE ã‚·ã‚¹ãƒ†ãƒ ã®ãƒ“ã‚¸ãƒã‚¹ä¾¡å€¤")
print("="*50)
print(f"âš¡ å‡¦ç†é€Ÿåº¦å‘ä¸Š: {efficiency_gain:.1f}%")
print(f"â±ï¸  1ãƒšãƒ¼ã‚¸ã‚ãŸã‚Šæ™‚çŸ­: {time_saving:.1f}ç§’")
print(f"ğŸ“Š ãƒ‡ãƒ¼ã‚¿åé›†ç²¾åº¦: {len(results['collected_data'])}ä»¶ / {results['total_screenshots']}ã‚¹ã‚¯ãƒªãƒ¼ãƒ³ã‚·ãƒ§ãƒƒãƒˆ")
print(f"\nğŸ’° æœˆé–“ã‚³ã‚¹ãƒˆå‰Šæ¸›åŠ¹æœï¼ˆ{pages_per_month}ãƒšãƒ¼ã‚¸å‡¦ç†ã®å ´åˆï¼‰:")
print(f"  ğŸ• æ™‚é–“çŸ­ç¸®: {monthly_time_saving_hours:.1f}æ™‚é–“")
print(f"  ğŸ’´ ã‚³ã‚¹ãƒˆå‰Šæ¸›: Â¥{monthly_cost_saving:,.0f}")

print("\nğŸ¯ AIç¤¾é•·ã®æˆ¦ç•¥çš„ä¾¡å€¤:")
print("  ğŸ”¹ 24æ™‚é–“ç„¡äººç¨¼åƒå¯èƒ½")
print("  ğŸ”¹ äººçš„ãƒŸã‚¹æ’é™¤")
print("  ğŸ”¹ ã‚¹ã‚±ãƒ¼ãƒ©ãƒ–ãƒ«ãªå‡¦ç†èƒ½åŠ›")
print("  ğŸ”¹ ãƒªã‚¢ãƒ«ã‚¿ã‚¤ãƒ å¸‚å ´ãƒ‡ãƒ¼ã‚¿åé›†")
print("  ğŸ”¹ 0å††ã§ã‚‚å®Ÿãƒ‡ãƒ¼ã‚¿ã§ä¾¡å€¤è¨¼æ˜")

# ROIè¨ˆç®—
development_investment = 500000  # é–‹ç™ºæŠ•è³‡é¡ï¼ˆå††ï¼‰
monthly_roi = (monthly_cost_saving / development_investment) * 100
payback_months = development_investment / monthly_cost_saving

print(f"\nğŸ“ˆ ROIåˆ†æ:")
print(f"  ğŸ’° æœˆæ¬¡ROI: {monthly_roi:.1f}%")
print(f"  ğŸ“… æŠ•è³‡å›åæœŸé–“: {payback_months:.1f}ãƒ¶æœˆ")
```
