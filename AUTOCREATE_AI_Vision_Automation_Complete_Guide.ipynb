## ğŸ” 2. OCRè§£æã‚·ã‚¹ãƒ†ãƒ  - Google Apps Script APIé€£æº

AUTOCREATEç‹¬è‡ªé–‹ç™ºã®OCRè§£æã‚·ã‚¹ãƒ†ãƒ ã‚’ä½“é¨“ã—ã¾ã™ã€‚
Google Apps Scriptï¼ˆGASï¼‰ã‚’ä½¿ç”¨ã—ãŸOCR APIã¨ã®é€£æºã«ã‚ˆã‚Šã€
ç”»åƒã‹ã‚‰æ–‡å­—ã‚’æŠ½å‡ºã—ã€ãƒ“ã‚¸ãƒã‚¹ä¾¡å€¤ã®ã‚ã‚‹æƒ…å ±ã«å¤‰æ›ã—ã¾ã™ã€‚

### ç‰¹å¾´
- **é«˜ç²¾åº¦OCR**: Google Vision APIãƒ™ãƒ¼ã‚¹
- **ãƒ“ã‚¸ãƒã‚¹ç‰¹åŒ–**: kinkaimasu.jpç­‰ã®å®Ÿã‚µã‚¤ãƒˆè§£æ
- **è‡ªå‹•æ”¹å–„ææ¡ˆ**: OCRçµæœã«åŸºã¥ãæ¬¡ã‚¢ã‚¯ã‚·ãƒ§ãƒ³ææ¡ˆ
- **å®Œå…¨è‡ªå‹•åŒ–**: äººé–“ã®ä»‹å…¥ä¸è¦

````python
def analyze_image_with_ocr(image_path, analysis_type="general"):
    """
    ç”»åƒã‚’OCRè§£æ
    Args:
        image_path (str): ç”»åƒãƒ•ã‚¡ã‚¤ãƒ«ãƒ‘ã‚¹
        analysis_type (str): è§£æã‚¿ã‚¤ãƒ—ï¼ˆgeneral, kinkaimasu, automationï¼‰
    Returns:
        dict: OCRè§£æçµæœ
    """
    try:
        # ç”»åƒã‚’Base64ã‚¨ãƒ³ã‚³ãƒ¼ãƒ‰
        with open(image_path, "rb") as img_file:
            base64_image = base64.b64encode(img_file.read()).decode('utf-8')
        
        # GAS OCR APIã«é€ä¿¡
        payload = {
            'image': base64_image,
            'fileName': os.path.basename(image_path),
            'type': analysis_type
        }
        
        print(f"ğŸ“¤ OCRè§£æé–‹å§‹: {os.path.basename(image_path)}")
        response = requests.post(autocreate_system.gas_ocr_api, data=payload, timeout=60)
        result = response.json()
        
        if result.get('success'):
            print("âœ… OCRè§£ææˆåŠŸï¼")
            return result
        else:
            print(f"âŒ OCRè§£æå¤±æ•—: {result.get('error', 'ä¸æ˜ãªã‚¨ãƒ©ãƒ¼')}")
            return result
            
    except Exception as e:
        print(f"ğŸ’¥ OCRè§£æã‚¨ãƒ©ãƒ¼: {str(e)}")
        return {"success": False, "error": str(e)}

def display_ocr_results(ocr_result):
    """OCRè§£æçµæœã‚’è¡¨ç¤º"""
    if not ocr_result.get('success'):
        print(f"âŒ è§£æå¤±æ•—: {ocr_result.get('error', 'ä¸æ˜ãªã‚¨ãƒ©ãƒ¼')}")
        return
    
    data = ocr_result.get('data', {})
    
    print("ğŸ“Š === OCRè§£æçµæœ ===")
    print(f"ğŸ¯ ä¿¡é ¼åº¦: {data.get('confidence', 0)}%")
    print(f"ğŸ“ ãƒ†ã‚­ã‚¹ãƒˆé•·: {len(data.get('ocrText', ''))} æ–‡å­—")
    
    # kinkaimasu.jpå°‚ç”¨åˆ†æ
    if data.get('isKinkamasuSite'):
        print("ğŸª kinkaimasu.jpã‚µã‚¤ãƒˆã‚’æ¤œå‡º")
        
        # é‡‘ä¾¡æ ¼æƒ…å ±
        gold_prices = data.get('goldPrices', [])
        if gold_prices:
            print("\nğŸ’° æ¤œå‡ºã•ã‚ŒãŸé‡‘ä¾¡æ ¼:")
            for i, price in enumerate(gold_prices[:3], 1):
                print(f"  {i}. {price.get('text', '')}")
        
        # ã‚·ã‚¹ãƒ†ãƒ æƒ…å ±
        system_info = data.get('systemInfo', {})
        if system_info:
            print("\nğŸ”§ ã‚·ã‚¹ãƒ†ãƒ æƒ…å ±:")
            print(f"  SSL: {'âœ…' if system_info.get('hasSSL') else 'âŒ'}")
            print(f"  WordPress: {'âœ…' if system_info.get('hasWordpress') else 'âŒ'}")
        
        # æ”¹å–„ææ¡ˆ
        improvements = data.get('improvements', [])
        if improvements:
            print("\nğŸš€ AIç¤¾é•·ã‹ã‚‰ã®æ”¹å–„ææ¡ˆ:")
            for i, imp in enumerate(improvements[:3], 1):
                print(f"  {i}. {imp.get('suggestion', '')} ({imp.get('priority', '')}å„ªå…ˆåº¦)")
    
    # æŠ½å‡ºãƒ†ã‚­ã‚¹ãƒˆã®è¡¨ç¤º
    ocr_text = data.get('ocrText', '')
    if ocr_text:
        print(f"\nğŸ“„ æŠ½å‡ºãƒ†ã‚­ã‚¹ãƒˆï¼ˆæœ€åˆã®300æ–‡å­—ï¼‰:")
        print("="*50)
        print(ocr_text[:300] + ('...' if len(ocr_text) > 300 else ''))
        print("="*50)

# ã‚µãƒ³ãƒ—ãƒ«ç”»åƒã‚’ä½œæˆã—ã¦OCRè§£æã‚’å®Ÿè¡Œ
print("ğŸ–¼ï¸ ã‚µãƒ³ãƒ—ãƒ«ç”»åƒã‚’ä½œæˆã—ã¦OCRè§£æã‚’å®Ÿæ¼”ã—ã¾ã™...")
sample_image = autocreate_system.create_sample_image()

if sample_image and os.path.exists(sample_image):
    print(f"\nğŸ” {os.path.basename(sample_image)} ã‚’OCRè§£æä¸­...")
    result = analyze_image_with_ocr(sample_image, "kinkaimasu")
    display_ocr_results(result)
else:
    print("âŒ ã‚µãƒ³ãƒ—ãƒ«ç”»åƒã®ä½œæˆã«å¤±æ•—ã—ã¾ã—ãŸ")
````
