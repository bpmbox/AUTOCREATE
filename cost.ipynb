## 2ï¸âƒ£ Gitãƒªãƒã‚¸ãƒˆãƒªã®ã‚³ã‚¹ãƒˆè¨ˆç®—é–¢æ•°ã®å®Ÿè£…

Gitå±¥æ­´ã‹ã‚‰é–‹ç™ºã‚³ã‚¹ãƒˆã‚’åˆ†æãƒ»è¨ˆç®—ã™ã‚‹é–¢æ•°ç¾¤ã‚’å®Ÿè£…ã—ã¾ã™ã€‚

````python
class GitCostAnalyzer:
    """Git ãƒªãƒã‚¸ãƒˆãƒªã®ã‚³ã‚¹ãƒˆåˆ†æã‚¯ãƒ©ã‚¹"""
    
    def __init__(self, repo_path=".", hourly_rate=5000):
        """
        åˆæœŸåŒ–
        
        Args:
            repo_path (str): Gitãƒªãƒã‚¸ãƒˆãƒªã®ãƒ‘ã‚¹
            hourly_rate (int): æ™‚é–“ã‚ãŸã‚Šã®å˜ä¾¡ï¼ˆå††ï¼‰
        """
        self.repo_path = Path(repo_path).resolve()
        self.hourly_rate = hourly_rate
        self.git_data = {}
        
        print(f"ğŸ“Š Git ã‚³ã‚¹ãƒˆåˆ†æå™¨ åˆæœŸåŒ–å®Œäº†")
        print(f"ğŸ“‚ ãƒªãƒã‚¸ãƒˆãƒª: {self.repo_path}")
        print(f"ğŸ’° æ™‚çµ¦å˜ä¾¡: Â¥{self.hourly_rate:,}/æ™‚é–“")
    
    def run_git_command(self, command):
        """Gitã‚³ãƒãƒ³ãƒ‰ã‚’å®Ÿè¡Œã—çµæœã‚’å–å¾—"""
        try:
            result = subprocess.run(
                command, 
                shell=True, 
                cwd=self.repo_path,
                capture_output=True, 
                text=True, 
                encoding='utf-8'
            )
            if result.returncode == 0:
                return result.stdout.strip()
            else:
                print(f"âŒ Git ã‚³ãƒãƒ³ãƒ‰ã‚¨ãƒ©ãƒ¼: {result.stderr}")
                return None
        except Exception as e:
            print(f"âŒ å®Ÿè¡Œã‚¨ãƒ©ãƒ¼: {e}")
            return None
    
    def get_commit_stats(self):
        """ã‚³ãƒŸãƒƒãƒˆçµ±è¨ˆã‚’å–å¾—"""
        print("ğŸ“ˆ ã‚³ãƒŸãƒƒãƒˆçµ±è¨ˆã‚’åˆ†æä¸­...")
        
        # ç·ã‚³ãƒŸãƒƒãƒˆæ•°
        total_commits = self.run_git_command("git rev-list --count HEAD")
        
        # ä»Šæœˆã®ã‚³ãƒŸãƒƒãƒˆæ•°
        this_month = self.run_git_command("git rev-list --count --since='1 month ago' HEAD")
        
        # ä»Šé€±ã®ã‚³ãƒŸãƒƒãƒˆæ•°
        this_week = self.run_git_command("git rev-list --count --since='1 week ago' HEAD")
        
        # ä»Šæ—¥ã®ã‚³ãƒŸãƒƒãƒˆæ•°
        today = self.run_git_command("git rev-list --count --since='1 day ago' HEAD")
        
        # ã‚³ãƒ³ãƒˆãƒªãƒ“ãƒ¥ãƒ¼ã‚¿ãƒ¼æ•°
        contributors = self.run_git_command("git shortlog -sn | wc -l")
        
        # æœ€åˆã¨æœ€å¾Œã®ã‚³ãƒŸãƒƒãƒˆæ—¥
        first_commit = self.run_git_command("git log --reverse --format='%ci' | head -1")
        last_commit = self.run_git_command("git log -1 --format='%ci'")
        
        self.git_data['commits'] = {
            'total': int(total_commits) if total_commits else 0,
            'this_month': int(this_month) if this_month else 0,
            'this_week': int(this_week) if this_week else 0,
            'today': int(today) if today else 0,
            'contributors': int(contributors) if contributors else 0,
            'first_commit': first_commit,
            'last_commit': last_commit
        }
        
        print(f"âœ… ç·ã‚³ãƒŸãƒƒãƒˆæ•°: {self.git_data['commits']['total']:,}ä»¶")
        print(f"ğŸ‘¥ ã‚³ãƒ³ãƒˆãƒªãƒ“ãƒ¥ãƒ¼ã‚¿ãƒ¼: {self.git_data['commits']['contributors']}äºº")
        
        return self.git_data['commits']

# GitCostAnalyzer ã®ã‚¤ãƒ³ã‚¹ã‚¿ãƒ³ã‚¹ä½œæˆ
analyzer = GitCostAnalyzer(repo_path=".", hourly_rate=5000)
print("ğŸš€ Git ã‚³ã‚¹ãƒˆåˆ†æå™¨ æº–å‚™å®Œäº†!")
````

````python
# GitCostAnalyzer ã‚¯ãƒ©ã‚¹ã«è¿½åŠ ãƒ¡ã‚½ãƒƒãƒ‰ã‚’å®Ÿè£…

def get_file_stats(self):
    """ãƒ•ã‚¡ã‚¤ãƒ«çµ±è¨ˆã‚’å–å¾—"""
    print("ğŸ“ ãƒ•ã‚¡ã‚¤ãƒ«çµ±è¨ˆã‚’åˆ†æä¸­...")
    
    # ç·ãƒ•ã‚¡ã‚¤ãƒ«æ•°
    total_files = self.run_git_command("git ls-files | wc -l")
    
    # ç·è¡Œæ•°
    total_lines = self.run_git_command("git ls-files | xargs wc -l | tail -1")
    
    # ãƒ•ã‚¡ã‚¤ãƒ«ã‚¿ã‚¤ãƒ—åˆ¥çµ±è¨ˆ
    file_types = self.run_git_command("git ls-files | sed 's/.*\\.//' | sort | uniq -c | sort -nr")
    
    self.git_data['files'] = {
        'total_files': int(total_files) if total_files else 0,
        'total_lines': total_lines,
        'file_types': file_types
    }
    
    print(f"ğŸ“„ ç·ãƒ•ã‚¡ã‚¤ãƒ«æ•°: {self.git_data['files']['total_files']:,}å€‹")
    return self.git_data['files']

def calculate_development_time(self):
    """é–‹ç™ºæ™‚é–“ã‚’æ¨å®š"""
    print("â° é–‹ç™ºæ™‚é–“ã‚’æ¨å®šä¸­...")
    
    commits = self.git_data.get('commits', {})
    total_commits = commits.get('total', 0)
    
    # 1ã‚³ãƒŸãƒƒãƒˆã‚ãŸã‚Šå¹³å‡30åˆ†ã¨ä»®å®š
    estimated_hours = total_commits * 0.5
    
    # ãƒ—ãƒ­ã‚¸ã‚§ã‚¯ãƒˆæœŸé–“ã‚’è¨ˆç®—
    if commits.get('first_commit') and commits.get('last_commit'):
        try:
            first = datetime.strptime(commits['first_commit'][:19], '%Y-%m-%d %H:%M:%S')
            last = datetime.strptime(commits['last_commit'][:19], '%Y-%m-%d %H:%M:%S')
            project_days = (last - first).days + 1
        except:
            project_days = 1
    else:
        project_days = 1
    
    self.git_data['time'] = {
        'estimated_hours': estimated_hours,
        'project_days': project_days,
        'avg_hours_per_day': estimated_hours / project_days if project_days > 0 else 0
    }
    
    print(f"â±ï¸ æ¨å®šé–‹ç™ºæ™‚é–“: {estimated_hours:.1f}æ™‚é–“")
    print(f"ğŸ“… ãƒ—ãƒ­ã‚¸ã‚§ã‚¯ãƒˆæœŸé–“: {project_days}æ—¥")
    
    return self.git_data['time']

def calculate_costs(self):
    """ã‚³ã‚¹ãƒˆã‚’è¨ˆç®—"""
    print("ğŸ’° ã‚³ã‚¹ãƒˆã‚’è¨ˆç®—ä¸­...")
    
    time_data = self.git_data.get('time', {})
    estimated_hours = time_data.get('estimated_hours', 0)
    
    # åŸºæœ¬äººä»¶è²»
    labor_cost = estimated_hours * self.hourly_rate
    
    # é–“æ¥è²»ï¼ˆåŠ´å‹™è²»ã®30%ã¨ä»®å®šï¼‰
    indirect_cost = labor_cost * 0.3
    
    # ç·ã‚³ã‚¹ãƒˆ
    total_cost = labor_cost + indirect_cost
    
    self.git_data['costs'] = {
        'labor_cost': labor_cost,
        'indirect_cost': indirect_cost,
        'total_cost': total_cost,
        'hourly_rate': self.hourly_rate,
        'estimated_hours': estimated_hours
    }
    
    print(f"ğŸ’µ äººä»¶è²»: Â¥{labor_cost:,.0f}")
    print(f"ğŸ¢ é–“æ¥è²»: Â¥{indirect_cost:,.0f}")  
    print(f"ğŸ’° ç·ã‚³ã‚¹ãƒˆ: Â¥{total_cost:,.0f}")
    
    return self.git_data['costs']

def analyze_full(self):
    """å®Œå…¨åˆ†æã‚’å®Ÿè¡Œ"""
    print("ğŸš€ å®Œå…¨ã‚³ã‚¹ãƒˆåˆ†æã‚’é–‹å§‹...")
    print("="*50)
    
    # å…¨åˆ†æã‚’å®Ÿè¡Œ
    self.get_commit_stats()
    self.get_file_stats()
    self.calculate_development_time()
    self.calculate_costs()
    
    print("="*50)
    print("âœ… ã‚³ã‚¹ãƒˆåˆ†æå®Œäº†!")
    
    return self.git_data

# GitCostAnalyzer ã‚¯ãƒ©ã‚¹ã«ãƒ¡ã‚½ãƒƒãƒ‰ã‚’è¿½åŠ 
GitCostAnalyzer.get_file_stats = get_file_stats
GitCostAnalyzer.calculate_development_time = calculate_development_time
GitCostAnalyzer.calculate_costs = calculate_costs
GitCostAnalyzer.analyze_full = analyze_full

print("ğŸ”§ GitCostAnalyzer ã‚¯ãƒ©ã‚¹æ‹¡å¼µå®Œäº†!")
````

````python
class GitHubIssueCreator:
    """GitHub Issueè‡ªå‹•ä½œæˆã‚¯ãƒ©ã‚¹"""
    
    def __init__(self, repo="bpmbox/AUTOCREATE"):
        self.repo = repo
        self.check_gh_cli()
    
    def check_gh_cli(self):
        """GitHub CLI ã®èªè¨¼çŠ¶æ…‹ã‚’ç¢ºèª"""
        try:
            result = subprocess.run(
                ["gh", "auth", "status"], 
                capture_output=True, 
                text=True
            )
            if result.returncode == 0:
                print("âœ… GitHub CLI èªè¨¼æ¸ˆã¿")
                return True
            else:
                print("âŒ GitHub CLI èªè¨¼ãŒå¿…è¦")
                print("ğŸ’¡ 'gh auth login' ã‚’å®Ÿè¡Œã—ã¦ãã ã•ã„")
                return False
        except FileNotFoundError:
            print("âŒ GitHub CLI ãŒã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«ã•ã‚Œã¦ã„ã¾ã›ã‚“")
            print("ğŸ’¡ https://cli.github.com/ ã‹ã‚‰ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«ã—ã¦ãã ã•ã„")
            return False
    
    def create_cost_issue(self, report_content, costs_data):
        """ã‚³ã‚¹ãƒˆãƒ¬ãƒãƒ¼ãƒˆã®Issueã‚’ä½œæˆ"""
        
        # UTF-8è¨­å®š
        env = os.environ.copy()
        env['PYTHONUTF8'] = '1'
        
        # Issue ã‚¿ã‚¤ãƒˆãƒ«
        total_cost = costs_data.get('total_cost', 0)
        date_str = datetime.now().strftime('%Y-%m')
        title = f"ğŸ“Š æœˆæ¬¡é–‹ç™ºã‚³ã‚¹ãƒˆãƒ¬ãƒãƒ¼ãƒˆ {date_str} - ç·ã‚³ã‚¹ãƒˆ Â¥{total_cost:,.0f}"
        
        # Issue æœ¬æ–‡ã‚’æº–å‚™ï¼ˆæ–‡å­—æ•°åˆ¶é™å¯¾å¿œï¼‰
        body = f"""# ğŸ“Š Gité–‹ç™ºã‚³ã‚¹ãƒˆãƒ¬ãƒãƒ¼ãƒˆ

{report_content[:2000]}

---

ğŸ¤– **è‡ªå‹•ç”Ÿæˆãƒ¬ãƒãƒ¼ãƒˆ** - {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}

ğŸ’° **ç·ã‚³ã‚¹ãƒˆ**: Â¥{total_cost:,.0f}
â±ï¸ **é–‹ç™ºæ™‚é–“**: {costs_data.get('estimated_hours', 0):.1f}æ™‚é–“
ğŸ’µ **æ™‚çµ¦**: Â¥{costs_data.get('hourly_rate', 0):,}/æ™‚é–“

ğŸ“‹ **ãƒ©ãƒ™ãƒ«**: cost-analysis, monthly-report, automation
"""
        
        try:
            # GitHub CLI ã§Issueä½œæˆ
            cmd = [
                "gh", "issue", "create",
                "--repo", self.repo,
                "--title", title,
                "--body", body,
                "--label", "cost-analysis,monthly-report,automation"
            ]
            
            print(f"ğŸš€ GitHub Issue ä½œæˆä¸­...")
            print(f"ğŸ“‹ ãƒªãƒã‚¸ãƒˆãƒª: {self.repo}")
            print(f"ğŸ“ ã‚¿ã‚¤ãƒˆãƒ«: {title}")
            
            result = subprocess.run(
                cmd,
                env=env,
                capture_output=True,
                text=True,
                encoding='utf-8'
            )
            
            if result.returncode == 0:
                issue_url = result.stdout.strip()
                print(f"âœ… Issueä½œæˆæˆåŠŸ!")
                print(f"ğŸ”— URL: {issue_url}")
                return issue_url
            else:
                print(f"âŒ Issueä½œæˆå¤±æ•—: {result.stderr}")
                return None
                
        except Exception as e:
            print(f"âŒ ã‚¨ãƒ©ãƒ¼: {e}")
            return None
    
    def create_summary_issue(self, analysis_results):
        """ã‚µãƒãƒªãƒ¼Issueã‚’ä½œæˆ"""
        
        commits = analysis_results.get('commits', {})
        costs = analysis_results.get('costs', {})
        time_data = analysis_results.get('time', {})
        
        # ç°¡æ½”ãªã‚µãƒãƒªãƒ¼
        summary_body = f"""# ğŸ¯ é–‹ç™ºã‚³ã‚¹ãƒˆã‚¯ã‚¤ãƒƒã‚¯ã‚µãƒãƒªãƒ¼

## ğŸ’° ã‚³ã‚¹ãƒˆæ¦‚è¦
- **ç·ã‚³ã‚¹ãƒˆ**: Â¥{costs.get('total_cost', 0):,.0f}
- **äººä»¶è²»**: Â¥{costs.get('labor_cost', 0):,.0f} 
- **é–“æ¥è²»**: Â¥{costs.get('indirect_cost', 0):,.0f}

## ğŸ“Š é–‹ç™ºçµ±è¨ˆ
- **ç·ã‚³ãƒŸãƒƒãƒˆ**: {commits.get('total', 0):,}ä»¶
- **é–‹ç™ºæ™‚é–“**: {time_data.get('estimated_hours', 0):.1f}æ™‚é–“
- **ãƒ—ãƒ­ã‚¸ã‚§ã‚¯ãƒˆæœŸé–“**: {time_data.get('project_days', 0)}æ—¥

## ğŸ² åŠ¹ç‡æŒ‡æ¨™
- **1ã‚³ãƒŸãƒƒãƒˆå˜ä¾¡**: Â¥{(costs.get('total_cost', 0) / max(commits.get('total', 1), 1)):,.0f}
- **æ™‚çµ¦åŠ¹ç‡**: Â¥{costs.get('hourly_rate', 0):,}/æ™‚é–“

---

ğŸ¤– è‡ªå‹•ç”Ÿæˆ: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
"""
        
        title = f"âš¡ é–‹ç™ºã‚³ã‚¹ãƒˆã‚µãƒãƒªãƒ¼ {datetime.now().strftime('%Y-%m-%d')}"
        
        try:
            cmd = [
                "gh", "issue", "create",
                "--repo", self.repo,
                "--title", title,
                "--body", summary_body,
                "--label", "cost-summary,automation"
            ]
            
            env = os.environ.copy()
            env['PYTHONUTF8'] = '1'
            
            result = subprocess.run(cmd, env=env, capture_output=True, text=True)
            
            if result.returncode == 0:
                issue_url = result.stdout.strip()
                print(f"âœ… ã‚µãƒãƒªãƒ¼Issueä½œæˆæˆåŠŸ!")
                print(f"ğŸ”— URL: {issue_url}")
                return issue_url
            else:
                print(f"âŒ ã‚µãƒãƒªãƒ¼Issueä½œæˆå¤±æ•—: {result.stderr}")
                return None
                
        except Exception as e:
            print(f"âŒ ã‚¨ãƒ©ãƒ¼: {e}")
            return None

# GitHub Issue ä½œæˆå™¨ã‚’åˆæœŸåŒ–
issue_creator = GitHubIssueCreator(repo="bpmbox/AUTOCREATE")
print("ğŸš€ GitHub Issue ä½œæˆå™¨ æº–å‚™å®Œäº†!")
````
